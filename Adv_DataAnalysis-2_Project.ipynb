#### 1) Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization

```python
vehicle_count = df['Brand'].value_counts().sort_values()
vehicle_count
```

```python
plt.style.use('classic')
plt.style.use('dark_background')
plt.figure(figsize=(8,4))
vehicle_count.plot(kind='bar', color='#efb1fa')
plt.title('Total Count of Vehicles by Brand')
plt.xlabel('Vehicle Brand')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

* Above plot is showing no of vehicle based on vehicles brand , there is maximum vehicle from volkswagon near about 80000

#### 2) What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis

```python
avg_prices = df.groupby(['Vehicle Type', 'Gearbox Type'])['Price'].mean().unstack()
avg_prices
```

```python
avg_prices.plot(kind='bar', figsize=(6,4),
                xlabel = ('Vehicle Type'),
                ylabel = ('Average Price'),
                title = ('Average Price of Vehicles by Type and Gearbox of vehicle'),
                legend = True)
```

* Above plot is showing manual gearbox from andere vehicle type having maximum avg price

#### 3) What is the marginal probability of private seller

```python
marginal_probability_private = len(df[df['Seller'] == 'privat']) / len(df['Seller'])

print("Marginal probability of 'private' seller in the 'seller' column is --->", marginal_probability_private)
```
