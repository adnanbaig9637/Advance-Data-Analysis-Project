#### 1) Can you tell me the Distribution of Vehicles based on Year of Registration with the help of a plot

```python
plt.style.use('classic')
plt.style.use('dark_background')
df['Year Of Registration'].plot(kind='hist',
                               color = 'blue',
                               grid = True,
                               figsize = (6,4),
                               title = 'Distribution of vehicles on Years Registration',
                               xlabel = 'Years')
```

* Above plot is showing the distribution of Vehicle based on Registraion Year , the maximum vehicles are having registration year 2000
* There less vehicles with old registration years.

#### 2) Create a plot based on the Variation of the price range by the vehicle type

```python
plt.style.use('classic')
plt.style.use('dark_background')
df.groupby(by='Vehicle Type')['Price'].mean().sort_values().plot(kind = 'bar',
                                                      color = '#fc0366',
                                                      figsize=(6,4),
                                                      ylabel='Prices',
                                                      title = 'Variation of Prices with vehicle type') 
```

* Above plot is showing there is high avg price in andere type of vehicle
* rest of vehicletypes are less prices

#### 3) Find out Total count of vehicles by type available on ebay for sale.As well as create a visualization for the client

```python
vehicle_count = df['Vehicle Type'].value_counts().sort_values()
vehicle_count
```

```python
plt.style.use('classic')
plt.style.use('dark_background')
plt.figure(figsize=(8,4))
vehicle_count.plot(kind='bar', color='skyblue')
plt.title('Total Count of Vehicles by Type on eBay')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Above plot is showing there is maximum and  near about 130000 vehicles are limousine type vehicle
* less vehicle are from andere type of vehicle
