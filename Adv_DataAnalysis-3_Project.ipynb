#### 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot

```python
avg_prices = df.groupby(['Fuel Type', 'Gearbox Type'])['Price'].mean().unstack()
avg_prices
```

```python
avg_prices.plot(kind='bar', figsize=(6,4),
                xlabel = ('Fuel Type'),
                ylabel = ('Average Price'),
                title = ('Average Price of Vehicles by Type and Gearbox of vehicle'),
                legend = True,
                color = ('#e6e273','#fc847c'))
```

* Above plot is showing manual gearbox and andere fuel type having maximum avg price

#### 3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
avg_power = df.groupby(['Vehicle Type', 'Gearbox Type'])['Power PS'].mean().unstack()
avg_power
```

```python
avg_power.plot(kind='bar', figsize=(6,4),
                xlabel = ('Vehicle Type'),
                ylabel = ('Average Power'),
                title = ('Average Price of Vehicles by Type and Gearbox of vehicle'),
                legend = True,
                color = ('#e6e273','#fc847c'))
```

*  Above plot is showing automatic gearbox from coupe vehicle type having maximum avg power more than 200 power ps

#### 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
avg_prices = df.groupby(['Brand', 'Vehicle Type'])['Price'].mean().reset_index()
```

```python
avg_prices
```

```python
pivot = avg_prices.pivot_table(index='Brand', columns='Vehicle Type', values='Price')

plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, cmap='YlGnBu',fmt='.2f',cbar_kws={'label': 'Average Price'})
plt.title('Heatmap for Average Price of Vehicles by Brand and Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Brand')
plt.yticks(rotation=0)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
